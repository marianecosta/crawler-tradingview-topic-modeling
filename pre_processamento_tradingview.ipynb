```python
import pandas as pd
import nltk
import string
from nltk.stem import PorterStemmer
```

```python
df = pd.read_csv('database-tradingview.csv')
display(df)
```

Primeira Etapa - Conversão da coluna Text de maiusculo para minusculo

```python
df['Content']=df['Content'].str.lower()
```

Segunda Etapa - Remove # e @

```python
df['Content'] = df['Content'].str.replace(r'[#@]', '', regex=True)
```

Terceira Etapa - Reduzir repetições a no máximo duas ocorrencias e intensificações

Quarta Etapa - Remove urls http ou www

```python
#Defini a expressão regular para URLs
padrao_url = re.compile(r'http\S+|www.\S+')

#Remove URLs da coluna "Content"
df['Content'] = df['Content'].str.replace(padrao_url, '', regex=True)
```

Quinta Etapa - Remoção de números e pontuações

```python
def preProcess(text):
    if isinstance(text, str):
        text = re.sub(r'[-./?!,":;()\']','',text)
        return text
    else:
        return text

df['Content'] = df['Content'].apply(preProcess)
```

Sexta Etapa - Remove tokens numericos

```python
padrao_numerico = re.compile(r'\b\d+\b|\b\d+\.\d+\b')

#Remove tokens numéricos da coluna "Text"
df['Content'] = df['Content'].str.replace(padrao_numerico, '', regex=True)
```

Sétima Etapa - Remove todas as palavras menores que 3 caracteres

```python
padrao_palavra = r'\b\w{1,5}\b'

#Remove palavras menores que 3 caracteres da coluna "Content"
df['Content'] = df['Content'].apply(lambda x: re.sub(padrao_palavra, '', x) if isinstance(x, str) else x)

display(df)
```

```python
import pandas as pd
from nltk.corpus import stopwords
stop_words = set(stopwords.words('portuguese'))

def remove_stop(text):
    if isinstance(text, str):
        return ' '.join(word for word in text.split() if word not in stop_words)
    else:
        return ''

df['Content'] = df['Content'].apply(remove_stop)

display(df)
```

Stemming

```python
stemmer = PorterStemmer()

df['Content'] = df['Content'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split()]))

display(df)
```
