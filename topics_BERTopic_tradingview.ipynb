```python
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import re
```

```python
comments_pd = pd.read_csv("base-portugues.csv")
```

```python
comments_pd['Content'] = comments_pd['Content'].astype(str)
commentslist = comments_pd['Content'].tolist()
```

**Filtrar por um ativo específico**





```python
comments_pd['Content'] = comments_pd['Content'].astype(str)
btcusd_comments = comments_pd[comments_pd['Ticker'] == 'BTCUSDT.P']
commentslist = btcusd_comments['Content'].tolist()
```

```python
model = BERTopic(
    language='portuguese', calculate_probabilities=True,
    verbose=True
)
topics, probs = model.fit_transform(commentslist)
```

```python
freq = model.get_topic_info()
freq.head(30)
```

```python
a_topic = freq.iloc[1]['Topic']
model.get_topic(a_topic)
```

```python
# Gráfico interativo para visualizar quais tópicos são semelhantes
model.visualize_topics()
```

```python
#Gráfico interativo das 5 principais palavras em cada tópico
model.visualize_barchart(top_n_topics=20)
```

```python
#Verifique o dendograma quanto à similaridade de tópicos
model.visualize_hierarchy(top_n_topics=30)
```
